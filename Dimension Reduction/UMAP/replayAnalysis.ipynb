```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import umap.umap_ as umap
import plotly.express as px
import numpy as np
```

```python
d = pd.read_csv(r"replayData.csv")
```

```python
data = d.dropna()
allGoodCol = data.columns
allGoodCol = list(allGoodCol)
allGoodCol = allGoodCol[2::]
```

```python
data
```

```python
reducer = umap.UMAP(n_neighbors = 5, 
                    metric = 'cosine', 
                    output_metric = 'euclidean',
                    learning_rate = 1.0, 
                    init = 'spectral', 
                    min_dist = 0.1, 
                    spread = 1.0, 
                    repulsion_strength = 1.0,
                    negative_sample_rate = 5,
                    target_metric = 'categorical', 
                    dens_lambda = 2.0, 
                    dens_frac = 0.3,
                    dens_var_shift=0.1,
                    n_components = 3)

parsedData = data[allGoodCol].values

embedding = reducer.fit_transform(parsedData)
embedding.shape
```

```python
df = pd.DataFrame(embedding, columns=['x1', 'x2', 'x3'])
df["binEdges"] = data.binEdges
df["id"] = data.id
df["order"] = data.order
```

```python
fig = px.scatter_3d(df, x='x1', y='x2', z='x3',
              color='order', opacity=0.9)

fig.update_traces(marker_size=2)

# tight layout
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```
